```python
# Music Script
local Music = game.Workspace.Game:WaitForChild("Music")
local Sound = script.Parent:WaitForChild("Sound")
local Forest = Sound.Forest
local Beach = Sound.Beach

local function EndMusic(touched, endMusic)
	if touched.Parent:FindFirstChild("Humanoid") then
		endMusic:Stop()
		print("Music ".. endMusic.Name .. " has stopped.")
	end
end

local function PlayMusic(touched, startMusic, endMusic)
	if touched.Parent:FindFirstChild("Humanoid") and startMusic.Playing == false then
		EndMusic(touched, endMusic)
		startMusic:Play()
		startMusic.Looped = true
		print("Music ".. startMusic.Name .. " is playing.")
	end
end

for key, object in pairs(Music:GetChildren()) do
	if object:IsA("Part") and object.Name == "Part1" then
		object.Touched:Connect(function(player)
			PlayMusic(player, Forest, Beach)
		end)
	elseif object:IsA("Part") and object.Name == "Part2" then
		object.Touched:Connect(function(player)
			PlayMusic(player, Beach, Forest)
		end)
	end
end
```

```python
# CHAT gpt attempt
local Music = game.Workspace.Game:WaitForChild("Music")
local Sound = script.Parent:WaitForChild("Sound")
local Forest = Sound.Forest
local Beach = Sound.Beach

local globalSound = Instance.new("Sound")
globalSound.Name = "GlobalMusic"
globalSound.Parent = game.Workspace
globalSound.SoundId = ""
globalSound.Volume = 0
globalSound.Looped = true
globalSound:Play()

local function EndMusic(player, startMusic)
	local sound = player.PlayerGui:FindFirstChild(startMusic.Name)
	if sound and sound.SoundId == startMusic.SoundId then
		sound:Stop()
		sound:Destroy()
		print("Music ".. startMusic.Name .. " has stopped for player "..player.Name..".")
	end
end


local function PlayMusic(player, startMusic, endMusic)
	if globalSound.SoundId ~= startMusic.SoundId then
		globalSound.SoundId = startMusic.SoundId
		globalSound.Volume = startMusic.Volume
	end
	local sound = player.PlayerGui:FindFirstChild(startMusic.Name)
	if player.Character and player.Character:FindFirstChild("Humanoid") then
		if not sound or sound.SoundId ~= globalSound.SoundId then
			EndMusic(player, endMusic)
			sound = Instance.new("Sound")
			sound.Name = startMusic.Name
			sound.Parent = player.PlayerGui
			sound.SoundId = globalSound.SoundId
			sound.Volume = globalSound.Volume
			sound.Looped = true
			sound:Play()
			print("Music ".. startMusic.Name .. " is playing for player "..player.Name..".")
		end
	else
		EndMusic(player, startMusic)
	end
end

for key, object in pairs(Music:GetChildren()) do
	if object:IsA("Part") and object.Name == "Part1" then
		object.Touched:Connect(function(otherPart)
			local player = game.Players:GetPlayerFromCharacter(otherPart.Parent)
			if player then
				PlayMusic(player, Forest, Beach)
			end
		end)
		object.TouchEnded:Connect(function(otherPart)
			local player = game.Players:GetPlayerFromCharacter(otherPart.Parent)
			if player then
				EndMusic(player, Forest)
			end
		end)
	elseif object:IsA("Part") and object.Name == "Part2" then
		object.Touched:Connect(function(otherPart)
			local player = game.Players:GetPlayerFromCharacter(otherPart.Parent)
			if player then
				PlayMusic(player, Beach, Forest)
			end
		end)
		object.TouchEnded:Connect(function(otherPart)
			local player = game.Players:GetPlayerFromCharacter(otherPart.Parent)
			if player then
				EndMusic(player, Beach)
			end
		end)
	end
end
```
